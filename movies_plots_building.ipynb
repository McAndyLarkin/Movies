```python
import matplotlib.pyplot as plt
```

```python
# plot/numpy

fig = plt.figure(figsize=(10, 6))

axes = fig.add_axes([0,0,1,1]) # создаем прямоуголник для построение графиков в виде списка [left, bottom, width, height]

axes.plot( xline, xline**2, 'r') # добавить первую кривую на холст красного цвета
axes.plot(xline, np.power(2, xline), 'b*--') # добавить вторую крикую на холст синего цвета с маркерами другого типа
axes.plot(xline, np.exp(xline), 'g--') # добавить вторую крикую на холст синего цвета с маркерами другого типа

axes.set_xlabel('x') # добавить название оси Х
axes.set_ylabel('y') # добавить название оси Y
axes.set_title('Название графика') # добавить название всего графика
axes.annotate("Это график", xy=(1. , 2.), xytext=(2.5, 3.), arrowprops={"facecolor":"grey", "shrink": 0.1})
axes.legend([r'$x^2$', r'$2^x$', r'$e^x$'], loc=0) # добавить легенду

# plt.show() # вывести график на экран
```

```python
import pandas as pd 

students = {'name':  ["Саша", "Маша", "Даша", "Аня", "Ваня", "Даня", "Саша1", "Маша1", "Даша1", "Аня1", "Ваня1", "Даня1", "Саша2", "Маша2", "Даша2", "Аня2", "Ваня2", "Даня2", "Саша3", "Маша3", "Даша3", "Аня3", "Ваня3", "Даня3"], 
            'age':   [    18,     19,     20,    21,    22,     23,      22,      18,      17,     19,     19,     20,      18,      19,      20,     20,      16,     19,      17,      21,      22,     20,     17,      16],
            'sex':   [   'M',    "F",    "F",   "F",   "M",    "M",     'F',     "M",     "F",    "M",    "M",    "F",     'M',     "F",     "M",    "F",     "M",    "F",     'F',     "F",     "F",    "M",    "M",     "M"],
            'height':[   1.8,   1.6,    1.55,  1.69,  1.75,    2.1,     1.6,    1.84,     1.95,   1.7,   1.62,   1.87,    1.84,    1.69,    1.51,   1.67,    1.45,    2.0,    1.65,     1.74,   1.64,   1.87,   1.54,    1.97],
        'profession':[   'D',   'S',     'S',    'L',   'D',   'E',     'E',     'D',      'S',    'D',   'S',    'E',     'S',     'L',     'L',    'D',     'L',    'E',     'S',      'S',    'L',    'E',    'L',     'D']
      }
professions = {
    "D": "Doctor",
    "E": "Engineer",
    "S": "Scientist",
    "L": "Jobless"
}
sex = {
    "F": "Female",
    "M": "Male"
}
students = pd.DataFrame(data=students)
students.profession = students.profession.apply(professions.get)
students.sex = students.sex.apply(sex.get)
students
```

```python
# hist/pandas
fig = plt.figure()
axes = fig.add_axes([0.0, 0.0, 1.0, 1.0])
males = students[students["sex"] == "Male"]
females = students[students["sex"] == "Female"]
cols = max(len(males), len(females)) #  количество столбцов

axes.hist(males['age'], bins=cols, alpha=0.6, label='Мужчины') # добавляем на холст гистограмму распределения возрастов среди мужчин
axes.hist(females['age'], bins=cols, alpha=0.6, label='Женщины') # добавляем на холст гистограмму распределения возрастов среди женщин

axes.legend() # строим легенду
axes.set_xlabel('Возраст', fontsize=18) 
axes.set_ylabel('Количество', fontsize=18)
axes.set_title('Распределение возрастов по полу человека', fontsize=18)

# plt.show() # Чтобы не пропадало окно вывода графика
```

```python
# pie/pandas
students.groupby('age').size().plot.pie(y='age', figsize=(5, 5))
```

```python
students_gr_by_sex_prof = students.groupby(["profession", "sex"]).size()
students_gr_by_sex_prof
```

```python
studs_by_prof_by_sex = students.groupby("profession").apply(lambda x: x.groupby("sex").size())
studs_by_prof_by_sex
```

```python
studs_by_prof_by_sex.plot.pie(subplots=True)
```

```python
studs_by_prof_by_sex = students.groupby("profession").apply(lambda x: list(x.groupby("sex").size()))
studs_by_prof_by_sex
```

```python
studs_by_prof_by_sex_list = np.array(studs_by_prof_by_sex.to_list())
studs_by_prof_by_sex_list
```

```python
# nested pie
fig, ax = plt.subplots()

size = 0.3

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(4)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10, 13, 14]))

ax.pie(studs_by_prof_by_sex_list.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels = list(studs_by_prof_by_sex.keys()))

ax.pie(studs_by_prof_by_sex_list.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels =["F", "M"]*4)

ax.set(title="Students grouped by 'profession' and 'sex'")
# plt.show()
```

```python
path_to_my_datasets = "./my_datasets"

movies = pd.read_csv(f'{path_to_my_datasets}/prepared_movies.csv')
movies
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'], format='%Y-%m-%d')
movies["release_date"]
```

```python
movies[movies.budget > 0].plot.line(title="Зависимость бюджета фильма от его даты производства", 
                            x='release_date', y='budget')
```

```python
movies[movies.budget > 0].plot.line(title="Зависимость бюджета фильма от его даты производства", 
                            x='release_date', y='budget', logy=True)
```

```python
movies[movies.budget > 0].sort_values(by=['budget'], inplace=False)
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(movies.corr());
plt.title("Корреляции между разными характеристиками фильмов");
```

```python
films_to_counries = pd.read_csv(f'{path_to_my_datasets}/films_to_counries.csv')
films_to_languages = pd.read_csv(f'{path_to_my_datasets}/films_to_languages.csv')
films_to_genres = pd.read_csv(f'{path_to_my_datasets}/films_to_genres.csv')
films_to_companies = pd.read_csv(f'{path_to_my_datasets}/films_to_companies.csv')
```

```python
films_to_counries = films_to_counries.rename(columns={"id": "movie_id"})
films_to_counries
```

```python
films_to_languages = films_to_languages.rename(columns={"id": "movie_id"})
films_to_languages
```

```python
films_to_companies
```

```python
estimated_moviews = movies.merge(films_to_counries.groupby("movie_id").size().reset_index(name="countries_count"), \
                                 on="movie_id")
estimated_moviews = estimated_moviews.merge(films_to_languages.groupby("movie_id").size().reset_index(name="languages_count"), \
                                 on="movie_id")
estimated_moviews = estimated_moviews.merge(films_to_companies.groupby("movie_id").size().reset_index(name="companies_count"), \
                                 on="movie_id")
estimated_moviews
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(estimated_moviews.corr());
plt.title("Корреляции между разными характеристиками фильмов");
```

```python
movies_grouped_by_max_revenue_per_year = movies[(movies.budget > 0) & (movies.revenue > 0)].groupby(lambda index: movies.release_date.loc[index].year)\
            .aggregate({'revenue':'max','budget':'max'})
```

```python
# plt.figure(figsize=(15,10))
# plt.xtickslabels(range(400, 550, 10), range(0, 150, 10))
bar_plot = movies_grouped_by_max_revenue_per_year[(movies_grouped_by_max_revenue_per_year.budget < 150000000) & (movies_grouped_by_max_revenue_per_year.revenue < 150000000)]\
    .plot.bar(title="Зависимость сборов фильма от его бюджета", y=['revenue', "budget"], rot=0)
bar_plot.grid()

for n, label in enumerate(bar_plot.get_xticklabels()):
    if n % 7 != 0:
        label.set_visible(False)

plt.savefig('./outputs/movies_grouped_by_max_revenue_per_year.png')
plt.savefig('./outputs/movies_grouped_by_max_revenue_per_year.pdf')
# bar_plot.set_xticks([0, 1, 2])
# bar_plot.set_xticklabels(d for d in ["1900", "","1950", "2000"])
# bar_plot.set_yticklabels(d for d in range(0, 3, 1))
```

```python
import pandas as pd 

students = {'name':  ["Саша", "Маша", "Даша"], 
            'age1':   [    1,     4,     7],
            'age2':   [    2,     5,     8],
            'age3':   [    3,     6,     9]
      }

students = pd.DataFrame(data=students)
students
```

```python
melted = pd.melt(students, id_vars=['name'], value_vars=['age1',"age2","age3"],
        var_name='del_it', value_name='age').drop(columns=["del_it"])
melted
```

```python
res = melted.groupby('name')["age"].apply(lambda x: pd.Series(x.values)).unstack()
res = res.rename(columns={i: f'age{i}' for i in range(res.shape[1])}).reset_index()
res
```
