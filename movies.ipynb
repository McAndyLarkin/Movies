```python
# Этот трюк повторять не нужно. Только если есть необходимость
import json
from collections import OrderedDict
from pandas.core.frame import DataFrame
import re 

"""
    Эта функция работает с таблицей (table:DataFrame), в которой колонка (column_name), 
    имеет cтроковые значения в формате json списка нескольких элементов 
    (например: 
        [
            {'iso_3166_1': 'US', 'name': 'United States of America'}, 
            {'iso_3166_1': 'RU', 'name': 'Russia'}
        ]
    ).
    Функция не меняет предоставленную таблицу (table:DataFrame).
    
    Данная функция преобразует значния всей колонки (column_name) в таблицу по типу:
    id		count	val0	val1
    123		2		US		RU
    
    Это первое возвращаемое значение функции. Всего их 2. О втором ниже.
    
    Название колонки id и ее значения (в примере: 123) повторяют переданную колонку (main_column) для исходной таблицы (table:DataFrame).
    Создается колонка count, которая содержит количество значений из json-списка, 
    а в следующих нумерованых колонках содержатся сами значения, 
    добытые из элементов json-списка по ключу (main_json_field).
    Количество нумерованых колонок вычисляется динамически.
    Названия нумерованных колонок задается с помощью (new_indexed_column_name).
    
    Второе возвращаемое значение функции представляет собой легенду распарсеных ячеек.
    Создается если передан флаг create_legend=True, чтобы данные не были утрачены.
    Легенда также представляет собой таблицу следующего вида:
    iso_3166_1	name
    US			United States of America
    RU			Russia
    
"""
def parse_json_column(table:DataFrame, column_name:str, main_column:str, main_json_field:str, 
                         new_indexed_column_name:str="val", create_legend:bool=False):
    legend_columns = None
    legend = None
    main_to_child = []
    actual_max_count = 0

    for i, row in table.iterrows():
        value = row[column_name]
        entities = []
        count = 0
        try:
#         if True:
            count = len(value)

            if create_legend and legend_columns is None and count > 0:
                legend_columns = []
                legend = set()
                entity = value[0]
                for col in entity:
                    legend_columns.append(col)

            for entity_json in value:
                if create_legend and legend is not None:
                    values = tuple(str(value) for value in entity_json.values())
                    legend.add(values)
                entities.append(entity_json[main_json_field])

            actual_max_count = max(actual_max_count, len(entities))
        except:
            print("Except", value)
        
        main_to_child.append([row[main_column], count, *entities])
        
    for row in main_to_child:
        while len(row) < actual_max_count:
            row.append(np.nan)

    columns = [f"{new_indexed_column_name}{i}" for i in range(actual_max_count)]
    main_to_child = pd.DataFrame(main_to_child, columns=[main_column, 'count', *columns])
    legend = pd.DataFrame(legend, columns=legend_columns)
    
    return main_to_child, legend

countries_aliases = [
    ("Cote D'Ivoire", "Cote D_Ivoire"), 
    ("Lao People's Democratic Republic", "Lao People_s Democratic Republic"), 
    ("'", '"')
]
countries_aliases = OrderedDict(countries_aliases)
films_to_counries, countries_legend = parse_json_column(movies, 
                                                        column_name='production_countries', 
                                                        main_column="id",
                                                        main_json_field = "iso_3166_1",
                                                        new_indexed_column_name="country",
                                                        create_legend=True)
```

```python
print("""Нужно решить проблему простоя ячеек. Исследуем несколько методов:
    1. Поскольку в нашем случае, чем меньше количество стран производителей (count), тем меньше ячеек задействовано (кроме 0), 
        мы можем разделить таблицу films_to_counries на две. В первой будут храниться фильмы, которые произведены малым
        количеством стран, соответственно количество ячеек там можно сократить, но количество сторк там будет большим. 
        Во второй таблице будет сильно меньше строк, но количество столбцов будет больше.
        Этот вариант не очень хороший по причине, которую мы увидим позже, а пока что в целях практики можно попробовать и его.
    2. Перекроить нашу таблицу так, чтобы в таблице с фильмами столбцы стран производителей были представлены строками. 
        Эту таблицу также стоит вынести отдельно от общей таблицы.
        
    Мы начнем с первого варианта, Рассмотрим его недочеты и придем ко второму.
""")
```

```python
# Разделение таблицы films_to_counries на две поменьше.
# Выберем порог, по которому будем делить таблицу. Будем выбирать самый выгодный с точки зрения экономии столбцов
benefit_threshold = 0
benefit_max = 0
for current_countries_count in range(1, max_countries):
    current_benefit = (max_countries - current_countries_count)*\
        len(films_to_counries[films_to_counries["count"] <= current_countries_count])
    if current_benefit > benefit_max:
        benefit_max = current_benefit
        benefit_threshold = current_countries_count
    
print(f"При пороге равном {benefit_threshold} будет достигнута наибольшая выгода в {benefit_max} ячеек!")
```

```python
# Поместим код по разбиению на таблицы в отдельную функцию, чтобы сократить количество дублирований.
# Ps Стоило это сделать сразу, но так примеры более доходчивы
def extruct_plural_map_df(table:DataFrame, column_name:str, main_column:str, main_json_field:str, 
                         create_legend:bool=False):
    new_indexed_column_name = f"val_{column_name}"
    map_df, legend = parse_json_column(table=table,
                                       column_name=column_name,  
                                       main_column=main_column,
                                       main_json_field = main_json_field,
                                       new_indexed_column_name=new_indexed_column_name,
                                       create_legend=create_legend)
    max_count = map_df["count"].max()

    plural_map = pd.melt(map_df, id_vars=[main_column], 
                         value_vars=[f"{new_indexed_column_name}{i}" for i in range(max_count)], 
                         var_name='del_it', value_name=main_json_field)
    plural_map = plural_map[plural_map[main_json_field].notnull()]
    plural_map = plural_map.drop(columns=["del_it"])

    table = table.drop(columns=[column_name])
    
    return (table, plural_map, legend)
```

```python
# И снова тот же фокус с жанрами
# пример значения ячейки
# genres -> [{'id': 28, 'name': 'Action'}, {'id': 80, 'name': 'Crime'}, 
#            {'id': 18, 'name': 'Drama'}, {'id': 53, 'name': 'Thriller'}]

movies.rename(columns = {'id':'movie_id'}, inplace = True)
movies, films_to_genres_plural, genres_legend = extruct_plural_map_df(table=movies, column_name="genres", 
                                                                      main_column="movie_id", main_json_field="id",
                                                                      create_legend=True)

some_film_id = '949'
some_film = movies[movies["movie_id"] == some_film_id].iloc[0]
some_film_genres = films_to_genres_plural[films_to_genres_plural["movie_id"] == some_film_id]

genres = [genre for genre in some_film_genres["id"].array]

full_genres_names = ", ".join([genres_legend[genres_legend["id"] == str(int(genre))].iloc[0]["name"] \
                        for genre in some_film_genres["id"].array])

print(f"""
    По классике. Фильм '{some_film["title"]}' в {some_film_genres["id"].count()} жанрах, среди которых:
    {full_genres_names}
""")
```
